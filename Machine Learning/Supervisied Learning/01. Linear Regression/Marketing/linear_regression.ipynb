```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LinearRegression
np.random.seed(0)
```

```python
df = pd.read_csv("advertising.csv")
df.head()
```

```python
sns.scatterplot(x ='TV', y = 'Sales', data = df)
```

```python
df = df.copy()
```

```python
# g = sns.FacetGrid()
```

```python
fig, ax = plt.subplots(1,3, figsize = (16, 6))
plt.style.use('seaborn')
ax[0].scatter(df.TV, df.Sales)
ax[0].set_xlabel('TV')

ax[1].scatter(df.Radio, df.Sales)
ax[1].set_xlabel('Radio')

ax[2].scatter(df.Newspaper, df.Sales)
ax[2].set_xlabel('Newspaper')
```

```python
x = df[['TV', 'Radio','Newspaper']]
y = df['Sales']
```

```python
y
```

### 00. To create any Machine Learning model in sklearn we are going to use 4 steps
### 01. Import the model
### 02. Instantiate/initialize the model
### 03. Fit the model
### 04. Predict

```python
from sklearn.linear_model import LinearRegression #importing
model = LinearRegression() #instantiate
model.fit(x, y) # fit the model, this is where OlS is applied #ols - ORDINARY LEAST SQUARES
```

```python
print(model.coef_)
print(model.intercept_)
```

```python
model.predict(np.array([180, 34, 200]).reshape(1, 3)) #TV - 180, #Radio - 34, #Newspapers - 200
```

```python
model.predict(np.array([180, 340, 20]).reshape(1, 3))  #TV - 180, #Radio - 340, #Newspapers - 20
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression #importing
model = LinearRegression() #instantiate
model.fit(X_train, y_train) # fit the model, this is where OlS is applied
```

```python
print(model.coef_)
print(model.intercept_)
```

```python
y_predic = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
score = r2_score(y_test, y_predic)
```
